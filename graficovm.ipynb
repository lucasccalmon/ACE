```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
oi = pd.read_csv('valor_monetario_ligas.csv', sep=';')
oi = oi.iloc[:,0:6]
ligas = oi['Ligas']
valores = oi.iloc[:,1:6]
oi.iloc[:, 1:] = oi.iloc[:, 1:].replace(',', '.', regex=True).astype(float)


oi
```

```python
plt.figure(figsize=(10, 6))
plt.style.use('ggplot')

# Configurando o fundo fora da área de plotagem
fig = plt.gcf()
fig.patch.set_facecolor('#1E2836')  # Cor de fundo fora da área de plotagem

ax = oi.plot(x='Ligas', kind='bar', stacked=False, figsize=(10, 6), color=['#FF5733', '#33FF57', '#3357FF', '#FF33A8', '#F3FF33'])

# Definindo a cor de fundo
ax.set_facecolor('#1E2836')
ax.grid(axis='x', visible=False)
# Configurações adicionais
ax.set_title('')
ax.set_xlabel('')
ax.set_ylabel('')
ax.legend().set_visible(True)
plt.tight_layout()
plt.savefig('desempenho_ligas.png', format='png')
# Exibindo o gráfico
plt.show()
```
