```python
participations = []

for phase, teams_list in championspais.items():
    for team in teams_list:
        participations.append((team, phase))

# Criar um DataFrame com as contagens de participação

df = pd.DataFrame(participations, columns=["País", "Phase"])


# Contar o número de participações por time em cada fase
participation_counts = pd.crosstab(df["País"], df["Phase"])
participation_counts = participation_counts.reindex(columns = championspais.columns)
participation_counts
columns_1920 = [col for col in participation_counts.columns if col.endswith("19/20")]
df_1920 = participation_counts[columns_1920]
```

```python


# Usando o dataframe já criado (df_sula)
df_champions_melted = championspais.melt(var_name='Fase', value_name='País')

# Remover valores nulos (times que não chegaram a certas fases)
df_champions_melted = df_champions_melted.dropna()

# Contagem de quantos times chegaram em cada fase por temporada
df_count = df_champions_melted.groupby(['Fase', 'País']).size().unstack().fillna(0)

# Plotar o gráfico de barras empilhadas
df_count.T.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='tab20')

# Adicionar título e rótulos aos eixos
plt.title('Participação dos Países por fase da Champions (2019-2023)')
plt.xlabel('País')
plt.ylabel('Número de Participações')
plt.legend(title='Fase', bbox_to_anchor=(1.05, 1), loc='upper left')

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
